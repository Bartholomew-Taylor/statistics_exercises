## terminology


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import viz # curriculum viz example code

np.random.seed(129)
```

```python
dice_distribution = stats.randint(1,7)
```

```python
die_distribution.rvs((5,5))
```

```python
#pmf for prob of specific value
die_distribution.pmf(3)
```

```python
#of a value and below
die_distribution.cdf(3)
```

```python
#of above a value
die_distribution.sf(4)
```

```python
## value associated with a probability
die_distribution.ppf(5/6)
```

```python
#inverse survival
#look up later
die_distribution.isf(1/3)
```

```python
n = 10_000

x = die_distribution.rvs(n)

plt.hist(x, bins=range(0, 9), align='left', width=1, edgecolor='black')
plt.title(f'Outcome of {n:,} Dice Rolls')
plt.ylim(1600, 1750)
```
